```python
unit_A = data["Unit A"]
unit_B = data["Unit B"]
```

```python
data.shape
```

```python
data.var()# two units does not follow same variance
```

```python
data.describe()
```

```python
p_value = stats.ttest_ind(unit_A,unit_B,equal_var=False)
p_value
```

```python
alpha=0.05
```

```python
if p_value[1] > alpha:
  print("Fail to Reject the null hypothesis cutlet have no significance difference in the diameter")
else:
    print("cutlet have  significance difference in the diameter")
```
